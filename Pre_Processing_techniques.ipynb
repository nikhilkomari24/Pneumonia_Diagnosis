## Gaussian Differencing Filter

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
image = cv2.imread('person5_bacteria_16.jpeg')

figure_size = 9 # the dimension of the x and y axis of the kernal.
blur = cv2.GaussianBlur(image ,(31, 31),0) #Obtaining the gaussian blured image of the original
FILTERED = 5*image - 4*blur   #subtracting the gaussian blured image from the original in a ratio 5:4 (No reason chosen maually after 3 trails)

plt.subplot(121), plt.imshow(image, cmap = 'gray'),plt.title('Original Image')

plt.subplot(122),plt.imshow(FILTERED, cmap = 'gray'),plt.title('Filtered image')

plt.show()


```

## Gamma Correction Filter

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('person5_bacteria_16.jpeg',0)

image = cv2.imread('person5_bacteria_16.jpeg', 0).astype(np.float32) / 255 #reading the image and implementing the gamma correction formulae

gamma = 0.5
corrected_image = np.power(image, gamma) 

plt.subplot(121), plt.imshow(img, cmap = 'gray'),plt.title('Original Image')

plt.subplot(122),plt.imshow(corrected_image, cmap ='gray'),plt.title('Gamma corrected Image')


plt.show()
```

## N - CLAHE Filter

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('person5_bacteria_16.jpeg',0)

# create a CLAHE object (Arguments are optional).

normalizedImg = np.zeros((800, 800))
normalizedImg = cv2.normalize(img,  normalizedImg, 0, 255, cv2.NORM_MINMAX)
#
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8)) #creating a clahe filter
cl1 = clahe.apply(normalizedImg)  #obtaining the clahe image for the normalized image 


plt.subplot(121), plt.imshow(img, cmap = 'gray'),plt.title('Original Image')

plt.subplot(122),plt.imshow(cl1, cmap ='gray'),plt.title('N - CLAHE image')


plt.show()
```
